```r
N <- rnorm(2000,2,2)
```

```r
mediana <- function (X){
  X <- sort(X)
  n <- length(X)
  if (length(X)%%2 == 0){
    return (1/2*(X[n/2]+X[n/2+1]))
  } else {return(X[(n+1)/2])}
}
```

```r
mediana(N)
median(N)
```

```r
kwartyl1 <- function(X){
  X <- sort(X)
  n = length(X)
  #X <- X[1:n/2]  
  return (mediana(X[1:n/2]))
}

kwartyl3 <- function(X){
  X <- sort(X)
  n <- length(X)
  index <- n/2 
  #X <- X[n/2:n]  
  return (mediana(X[index:n]))
}
```

```r
kwartyl1(N)
kwartyl3(N)
quantile(N)
```

```r
rozstep <- function(X){
  return (max(X)-min(X))
}
```

```r
rozstep(N)
range(N)
```

```r
rozstep_miedzykw <- function(X){
  return (kwartyl3(X)-kwartyl1(X))
}
```

```r
rozstep_miedzykw(N) 
IQR(N)
```

```r
wariancja <- function(X){
  n = length(X)
  return (1/(n-1)*sum((X-mean(X))^2))
}
```

```r
wariancja(N)
var(N)
```

```r
odchylenie_std <- function(X){
  return (wariancja(X)^(1/2))
}
```

```r
odchylenie_std(N)
sd(N)
```

```r
wsp_zmiennosci <- function(X){
  return (odchylenie_std(X)/mean(X) * 100)
}
```

```r
wsp_zmiennosci(N)
cv(N) #w R dziala i jest git
```

```r
x <- dane$price
```

```r
rednia_arytmetyczna <-function(dane){
    return(sum(dane)/length(dane))
} 
```

```r
print(rednia_arytmetyczna(x))
```

```r
rednia_harmoniczna <- function(dane){
    return(length(dane)/sum(1/dane)) 
}
```

```r
rednia_harmoniczna(x)
```

```r
rednia_geometryczna <- function(dane){
    return(prod(dane^(1/length(dane))))
}
```

```r
rednia_geometryczna(x)
```

```r
rednia_winsorowska <- function(dane, k){
    n <- length(dane)
    posortowane = sort(dane)
    koniec <- n-k-1
    pocztek <- k+1
    obcite = posortowane[pocztek:koniec]
    return(1/n*((k+1)*posortowane[k+1] + sum(obcite)+ (k+1)*posortowane[n-k]))
}
```

```r
rednia_winsorowska(x, 300)
```

```r
k <- seq(1,floor(length(x)/2),10)
```

```r
y_axis_wins <- numeric()
i <- 1
for (n in 1:length(k)) {
    y_axis_wins[i]=rednia_winsorowska(x,i)
    i <- i+1
  }
plot(k,y_axis_wins, main= "rednia winsorowska dla rnej liczby ucitych wartoci", xlab="k", ylab="rednia winsorowska", col = "yellow")
```

```r
rednia_ucinana <- function(dane, k){
    n <- length(dane)
    koniec <- n-k
    pocztek <- k+1
    posortowane = sort(dane)
    obcite = posortowane[pocztek:koniec]
    return(1/(n-2*k)*sum(obcite))
}
```

```r
rednia_ucinana(x, 100)
```

```r
y_axis_uc <- numeric()
i <- 1
for (n in 1:length(k)) {
    y_axis_uc[i]=rednia_ucinana(x,i)
    i <- i+1
  }
plot(k,y_axis_uc, main= "rednia ucinana dla rnej liczby ucitych wartoci", xlab="k", ylab="rednia ucinana", col = "pink")
```
