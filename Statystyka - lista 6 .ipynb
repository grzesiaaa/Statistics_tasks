```r
n1 <- 20
n2 <- 50
n3 <- 100
mi <- 2.1
sigma <- 0.2
alpha <- 0.02
```

```r
#dla znanej sigmy
zad4 <- function(n){
    wpadniecia <- numeric()
    for(k in 1:100){
        ilosc_trafien <- 0
        for(i in 1:1000){
            X <- rnorm(n, mi, sigma)
            lewa_strona <- mean(X)-qnorm(1-alpha/2)*(sigma/(n^(1/2)))
            prawa_strona <- mean(X)+qnorm(1-alpha/2)*(sigma/(n^(1/2)))
            if(lewa_strona <= mi & mi <= prawa_strona){
                ilosc_trafien <- ilosc_trafien + 1
            }
        wpadniecia[k] <- ilosc_trafien/1000
        }
    }
    return(wpadniecia)
    }
```

```r
boxplot(zad4(n1))
```

```r
boxplot(zad4(n2))
```

```r
boxplot(zad4(n3))
```

```r
#dla nieznanej sigmy
zad4_niezn <- function(n){
    wpadniecia <- numeric()
    for(k in 1:100){
        ilosc_trafien <- 0
        for(i in 1:1000){
            X <- rnorm(n, mi, sigma)
            S <- (1/(n-1)*sum((X-mean(X))^2))^(1/2)
            lewa_strona <- mean(X)-qt(1-alpha/2, n-1)*(S/(n^(1/2)))
            prawa_strona <- mean(X)+qt(1-alpha/2, n-1)*(S/(n^(1/2)))
            if(lewa_strona <= mi & mi <= prawa_strona){
                ilosc_trafien <- ilosc_trafien + 1
            }
        wpadniecia[k] <- ilosc_trafien/1000
        }
    }
    return(wpadniecia)
    }
```

```r
boxplot(zad4_niezn(n1))
```

```r
boxplot(zad4_niezn(n2))
```

```r
boxplot(zad4_niezn(n3))
```
